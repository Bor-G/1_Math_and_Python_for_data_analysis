## Представляю итог по заданию "Центральная предельная теорема своими руками".

Для работы выбрано Гамма-распределение: https://ru.wikipedia.org/wiki/Гамма-распределение, документацию по работе с ним в Python можно посмотреть здесь: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.gamma.html.

```python
#Первым делом я импортировал нужные библиотеки.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from statsmodels.distributions.empirical_distribution import ECDF
%matplotlib inline
```

```python
#Далее я задал распределение и сгенерировал из него выборку указанного в условии объёма.
gamma_rv = sts.gamma(5., loc = 0., scale = 0.3)
samples = gamma_rv.rvs(1000)
```

```python
#Затем была построена гистограмма выборки, также теоретическая плотность распределения.
#Параметры подобраны вручную с целью улучшения визуализации.
plt.hist(samples, bins=30, density=True, label='HIST')
x = np.linspace(0,6,100)
pdf = gamma_rv.pdf(x)
plt.plot(x, pdf, label='TPDF')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
#Для 1000 выборок я решил рассмотреть 5 значений n.
#Ниже приведены гистограммы распределений их выборочных средних.
#На графиках заметно приближение к нормальности при росте n.
for n in [5, 10, 50, 500, 1000]:
    samples1 = np.array([gamma_rv.rvs(size=n) for i in range(1000)])
    means = samples1.mean(axis=1)
    plt.hist(means, density=True)
    plt.title('Means: $X_{%s}$' % n)
    plt.xlabel('$x$')
    plt.ylabel('Fraction')
    plt.show()
```

Возвращаясь к материалам Википедии, отмечу, что формула Гамма-распределения имеет следующий вид (для x ≥ 0):

$$ f(x)=\frac{x^{k-1}e^{-x/\theta}}{\theta^k \Gamma(\alpha)} $$

Математическое ожидание и дисперсия Гамма-функции определяются следующим образом: 

$$ \mathbb{E}(X) = k\theta, \mathbb{D}(X) = k\theta^2 $$

Таким образом, ЦПТ для Гамма-функции выглядит следующим образом:

$$ \overline{X_n} \approx\sim N(k\theta, \frac{k\theta^2}{n}) $$

```python
#Для быстрого определения теоретических значений я воспользовался командой из спецификации функции:
#https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.gamma.html
gamma_rv.stats(moments='mv')
```

```python
#В итоге я совместно отобразил гистограму по выборке и плотность нормального распределения для каждого из 5 случаев.
#При построении графика плотности использованы описанные выше теоретические параметры для Гамма-функции.
for n in [5, 10, 50, 500, 1000]:
    samples1 = np.array([gamma_rv.rvs(size=n) for i in range(1000)])
    means = samples1.mean(axis=1)
    plt.hist(means, density=True, label='HIST')
    plt.title('Means of $X_{%s}$' % n)
    
    norm_rv = sts.norm(loc=1.5, scale=np.sqrt(0.45 / n))
    x = np.linspace(0,3,100)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf, label='NORM')
    plt.legend()
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    plt.show()
```

# Выводы по результатам анализа

Для наглядности было выбрано 5 значений n. По графикам очевидно следующее:
1. ЦПТ работает для Гамма-функции точно так же, как для распределений, рассмотренных в видеолекциях.
1. С ростом n повышается гладкость гистограммы выборочных средних, то есть она все лучше описывается нормальным распределением.
2. С ростом n гистограма становится более узкой, то есть точность нормальной апроксимации растет.
