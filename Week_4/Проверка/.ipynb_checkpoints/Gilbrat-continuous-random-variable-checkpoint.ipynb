#### Центральная предельная теорема своими руками

Для эксперимента возьмем логнормальное распределение(известное также как  A Gilbrat continuous random variable).
https://en.wikipedia.org/wiki/Log-normal_distribution

$$f(x) = \frac{1}{x}\frac{1}{\sigma\sqrt{2\pi}}\exp(-\frac{(\ln x-\mu)^2}{2\sigma^2})$$

Математическое ожидание и дисперсия:

$$E[X]=e^{\mu+\sigma^2/2}, \quad D[X]=(e^{\sigma^2}-1)e^{2\mu+\sigma^2}$$




При $\mu=0\quadи\quad\sigma^2=1$ :
$$f(x) = \frac{1}{x}\frac{1}{\sqrt{2\pi}}\exp(-\frac{(\ln x)^2}{2})$$

Математическое ожидание и дисперсия:
$$E[X]=\sqrt{e}$$

$$D[X]=(e-1)e$$


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Расчитаем теоретические математическое ожидание и дисперсию: 

```python
EX = np.sqrt(np.e)
DX = (np.e - 1)*np.e
print '[EX](teor) = ', EX
print '[DX](teor) = ', DX
```


Сгенерируем выборку объёма 1000 :


```python
gilbrat_rv = sts.gilbrat()
sample = gilbrat_rv.rvs(1000)
```

Построим гистограмму для выборки и график теоретической плотности распределения :

```python
nbins = 50 # number of bins:
plt.rcParams["figure.figsize"] = 10,5
plt.rcParams["axes.grid"] = True
plt.hist(sample, bins=nbins, normed=True,label='histogram')
x = np.linspace(sts.gilbrat.ppf(0.01), sts.gilbrat.ppf(0.99), 1000)
plt.plot(x, sts.gilbrat.pdf(x), label='theoretical')
plt.ylabel('$pdf(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
plt.show()
```

Распределение сильно скошено влево

Сгенерируем 1000 выборок объемом n=5 и получим выборку средних: 

```python
n = 5
means = [np.mean(gilbrat_rv.rvs(n)) for i in range(1000)]
```

Посчитаем значение параметров нормального распределения , к которому, согласно центральной предельной теореме, приближается распределение выборочных средних:

В соответствии с центральной предельной теоремой,  стандартное отклонение для апроксимирующего нормального распределения:

$\sigma=\sqrt{[D]/n}$

```python
sd = np.sqrt(DX/n)# standart deviation
print 'sd(n=%s) = ' % n, sd 
```

Построим график плотности вероятности расчитанного нормального распределение и гистограмму для выборочных средних для данного n:

```python
normal_rv = sts.norm(loc=EX, scale=sd) #normal 
x = np.linspace(0, 5, 1000)
plt.hist(means, bins=nbins, normed=True, label='histogram for n=%s' % n)
plt.plot(x, normal_rv.pdf(x), label='norm pdf')
plt.ylabel('fraction of means')
plt.xlabel('$mean(n=%s)$' % n)
plt.legend(loc='upper right')
plt.show()
```

Аналогично посчитаем значения  для нормального распределения при  n = 10, 50, 100 и построим графики плотности вероятности соответсвующих нормальных распределений:

```python
n = 10 
means = [np.mean(gilbrat_rv.rvs(n)) for i in range(1000)]
sd = np.sqrt(DX/n)# standart deviation
print 'sd(n=%s) = ' % n, sd 
normal_rv = sts.norm(loc=EX, scale=sd) #normal 
x = np.linspace(0, 5, 1000)
plt.hist(means, bins=nbins, normed=True, label='histogram for n=%s' % n)
plt.plot(x, normal_rv.pdf(x), label='norm pdf')
plt.ylabel('fraction of means')
plt.xlabel('$mean(n=%s)$' % n)
plt.legend(loc='upper right')
plt.show()
```

```python
n = 50
means = [np.mean(gilbrat_rv.rvs(n)) for i in range(1000)]
sd = np.sqrt(DX/n)# standart deviation
normal_rv = sts.norm(loc=EX, scale=sd) #normal 
x = np.linspace(0, 5, 1000)
plt.hist(means, bins=nbins, normed=True, label='means hist for n=%s' % n)
plt.plot(x, normal_rv.pdf(x), label='norm pdf')
plt.ylabel('fraction of means')
plt.xlabel('$mean(n=%s)$' % n)
plt.legend(loc='upper right')
plt.show()
```

```python
n = 300
means = [np.mean(gilbrat_rv.rvs(n)) for i in range(1000)]
sd = np.sqrt(DX/n)# standart deviation
print 'sd(n=%s) = ' % n, sd 
normal_rv = sts.norm(loc=EX, scale=sd) #normal 
x = np.linspace(0, 5, 1000)
plt.hist(means, bins=nbins, normed=True, label='histogram for n=%s' % n)
plt.plot(x, normal_rv.pdf(x), label='norm pdf')
plt.ylabel('fraction of means')
plt.xlabel('$mean(n=%s)$' % n)
plt.legend(loc='upper right')
plt.show()
```

##### Полученные результаты показали, что распределение выборочного среднего для случайной величины при увеличении объёма выборок приближается к нормальному распределению (в соответствии с центральной предельной теоремой) .
