# Центральная предельная теорема своими руками

Я выбрала распределение хи-квадрат, так как оно не очень похоже на нормальное.


Импортируем необходимые библиотеки:

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.stats as sts
from scipy.stats import chi2
import math
from math import sqrt
%matplotlib inline
```

Задаем постоянные и определяем распределение хи-квадрат:

```python
xm = 1
k = 3
E = 3
D = 6
n = 5

chi = sts.chi2(xm, k)
x = np.linspace(0, 5, 1000)
cdf = chi.cdf(x)

plt.plot(x, cdf)
```

Генерируем выборку объемом 1000 и строим гистограмму с плотностью распределения случайной величины:

```python
chi_sam = sts.chi2(3)
plt.hist(chi_sam.rvs(size = 1000), normed = True, bins = 150)
y = np.linspace(1, 5, 1000)
pdf = chi_sam.pdf(y)
#plt.xlim(1,5)
plt.plot(y, pdf)
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
```

Вычислим матожидание и дисперсию:

```python
Ex = chi.mean()
Dx = chi.std() ** 2
```

Создадим массив и запишем в него выборочные средние:

```python
array = []
count = 5
for i in range(1000):
    z = chi.rvs(count)
    array.append(sum(z) / len(z))
```

Считаем дисперсию нормального распределения:

```python
Dxn = Dx / count
normal = sts.norm(Ex, sqrt(Dxn)) 
pdf = normal.pdf(x)
```

Строим гистограмму для выборки размером 5:

```python
plt.plot(x, pdf)
plt.hist(array, density = True)
```

Повторим для выборки объемом 50:

```python
array = []
count = 50
for i in range(1000):
    z = chi.rvs(count)
    array.append(sum(z) / len(z))
    
Dxn = Dx / count
normal = sts.norm(Ex, sqrt(Dxn)) 
pdf = normal.pdf(x)

plt.plot(x, pdf)
plt.hist(array, density = True)
```

И для выборки объемом 100:

```python
array = []
count = 100
for i in range(1000):
    z = chi.rvs(count)
    array.append(sum(z) / len(z))
    
Dxn = Dx / count
normal = sts.norm(Ex, sqrt(Dxn)) 
pdf = normal.pdf(x)

plt.plot(x, pdf)
plt.hist(array, density = True)
```

По мере увеличения объема выборки гистограмма приближается к плотности распределения, дисперсия уменьшается, усиливается группировка вокруг среднего значения.
