# Центральная предельная теорема своими руками

Так как в задании было сказано что чем меньше распределение похоже на нормальное тем лучше, я выбрал бэта-распределение с такими параметрами, чтобы крайние значение имели большую вероятность.

Ниже представлен код для построения графика теоретической плотности распределения и гистограмма распределения выборки n=1000. Видно что общая тенденция совпадает. Графики не очень выровнены друг относително друга, посколько опция normed=True в моей версии plt не работала и dencity - самое похожее что я смог найти;(

```python
import scipy.stats as sts
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
%matplotlib inline

#Параметры a и b задают функцию плотности распределения и сильно влияют на график
alpha_value = 0.1
beta_value = 0.12
x = np.linspace(0, 1, 1002)
beta = sts.beta(alpha_value,beta_value)


cdf = beta.pdf(x)

count = 1000
y = beta.rvs(count)
plt.xlabel('Случайная величина')
plt.step(x, cdf)
plt.hist(y, density=True)
print('Alpha = ', alpha_value, ' beta = ', beta_value)
```

Далее я вычислил матожидание и дисперсию для функции распределения, сргласно википедии эти значения вычисляются из альфа и бэта, заданных при создании самого бэта-распределения. В моем случае было
alpha_value = 0.1
beta_value = 0.12
Сохраним матожидание как me и дисперсию как dis. Для того чтобы рассчитать дисперсию нормального распределения по центральной предельной теореме нужно разделить значение dis на количесто элеменов в выборке.
Это происходит в строке 
dis_norm = dis / count
Далее для выборок размеров 5, 10, 30 и 50 построим графики теоретической плотности и гистограмму с фактическим распределением.

```python
EX = beta.mean()
DX = beta.std() ** 2


#Сохраним в массив m выборочне средние из count элементов для 1000 выборок
m = []
count = 5
for i in range(1000):
    y = beta.rvs(count)
    m.append(sum(y) / len(y)) 
             
DXn = DX / count #разделим дисперсию на количество элементов в выборке в соотвествии с центральной предельное теоремой
norm_rv = sts.norm(EX, sqrt(DXn)) #функиця принимает среднеквадратическое отклонение, поэтому вычислим корень из дисперсии
pdf = norm_rv.pdf(x)

plt.xlabel('Случайная величина')
plt.plot(x, pdf)
plt.hist(m, density=True)

print('EX = ', EX, 'DXn = ', DXn)
```

```python
m = []
count = 10
for i in range(1000):
    y = beta.rvs(count)
    m.append(sum(y) / len(y)) 
             
DXn = DX / count
norm_rv = sts.norm(EX, sqrt(DXn))
pdf = norm_rv.pdf(x)

plt.xlabel('Случайная величина')
plt.plot(x, pdf)
plt.hist(m, density=True)

print('EX = ', EX, 'DXn = ', DXn)
```

```python
m = []
count = 30
for i in range(1000):
    y = beta.rvs(count)
    m.append(sum(y) / len(y)) 
             
DXn = DX / count
norm_rv = sts.norm(EX, sqrt(DXn))
pdf = norm_rv.pdf(x)

plt.xlabel('Случайная величина')
plt.plot(x, pdf)
plt.hist(m, density=True)

print('EX = ', EX, 'DXn = ', DXn)
```

```python
m = []
count = 50
for i in range(1000):
    y = beta.rvs(count)
    m.append(sum(y) / len(y)) 
             
DXn = DX / count
norm_rv = sts.norm(EX, sqrt(DXn))
pdf = norm_rv.pdf(x)

plt.xlabel('Случайная величина')
plt.plot(x, pdf)
plt.hist(m, density=True)
print('EX = ', EX, 'DXn = ', DXn)
```

Можно заметить что при зрачении n=30 гистограмма приближается к теоретической плотности распределения, дальнейшее увеличение n до 50 не особо
улучшает ситуацию. Также с увеличением элементов в распределени все сильнее группируется вокруг среднего значения, дисперсия уменьшается.
