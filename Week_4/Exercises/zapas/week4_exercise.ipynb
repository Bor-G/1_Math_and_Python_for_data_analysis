# Density against histogram

Let's import all the libraries we need first:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from math import *
%matplotlib inline
```

Now, we can initialize a wrapped Cauchy random variable and get some samples.

```python
wrapcauchy_rv = sts.wrapcauchy(0.3, 5)
samples = wrapcauchy_rv.rvs(1000)
```

Time to draw a histogram and a theoretical density function together:

```python
plt.hist(samples, bins=20, normed=True, label='Histogram', color='gray')
x = np.linspace(4, 12, 1000)
pdf = wrapcauchy_rv.pdf(x)
plt.plot(x, pdf, label='Theoretical pdf', color='red', linewidth=3)
plt.xlabel('$x$')
plt.legend()
plt.show()
```

# The cental limit theorem
We will need theoretical values for the mean and standard deviation of the distribution, so let's compute them now. Luckily enough, we have standard functions provided by SciPy to get them.

```python
mean = wrapcauchy_rv.mean()
standard_dev = wrapcauchy_rv.std()
```

Now we want to estimate the distribution of the sample mean values for different $N$ from ${5, 10, 50}$. Let's first write a function which generates 1000 samples of size $N$, then computes their sample mean values and returns these mean values as a result.

```python
def get_sample_mean_values(N, result_size):
    result = np.ndarray(result_size)
    for i in xrange(result_size):
        sample = wrapcauchy_rv.rvs(N)
        result[i] = sum(sample) / N
    return result
```

As mentioned in the task, we now want to see how close the distribution of sample mean values is to the corresponding normal distribution (based on the central limit theorem). Let's do a few examples.

## $N = 2, resultsize=1000$

```python
N, result_size = 2, 1000
sample_means = get_sample_mean_values(N, result_size)
norm_rv = sts.norm(mean, standard_dev / sqrt(N))
```

```python
plt.hist(sample_means, bins=20, normed=True, label='Histogram', color='gray')
x = np.linspace(4, 12, 1000)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='Theoretical pdf', color='red', linewidth=3)
plt.xlabel('$x$')
plt.legend()
plt.show()
```

## $N = 5, result size = 1000$
