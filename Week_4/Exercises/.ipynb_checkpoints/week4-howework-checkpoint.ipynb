```python
# распределение релея
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline

# 100 точек дла графика по оси Х
x = np.linspace(rayleigh.ppf(0), rayleigh.ppf(0.999), 100)

# выборка значения для распределения Релея, 1000 значений
from scipy.stats import rayleigh
r = rayleigh.rvs(size=1000)

# Гистограмма по распределению Релея из 20 столбцов
plt.hist(r, bins = 20, normed=True, facecolor='g', alpha=0.3, label='rayleigh histogram')

# График плотности распределения ( красный  ) probability density function (pdf)
plt.plot(x, rayleigh.pdf(x),'r-', lw=2, alpha=0.6, label='rayleigh pdf')

# оформление графика
plt.ylabel('$rayleigh$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
plt.title("Rayleigh")
plt.axis([0, 3.5, 0, 0.8])

plt.show()

```

```python

# функция, рисующая графики исходя из n - объема выборки
def show_average(n):
    experiments = 1000
    i = 0
    
    distr = np.empty(experiments)
    while i < experiments:
        r = rayleigh.rvs( size = n )
        distr[i] = np.mean(r)
        i = i + 1
    
    # гистограмма средних значений
    plt.hist(distr, bins= 30, normed=1, facecolor='r', alpha=0.3 )
    plt.axis([0, 2.5, 0, 5])
    
    # нормальное распределение ( теоретическое )
    norm_rv = sts.norm(rayleigh.mean(), math.sqrt(rayleigh.var() / n) )
    x = np.linspace(0,3,100)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf , label='theoretical norm')
    
    
    plt.xlabel("average for n = " + str(n))
    plt.legend(loc='upper right')
    plt.grid(1)
    plt.show()
    
show_average(5)
show_average(10)
show_average(50)
```
