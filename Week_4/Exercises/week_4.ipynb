Подключаем необходимые пакеты:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Сгенерируем выборку объёма 1000 из распределения хи-квадрат $χ_2^2$:

```python
chi2_rv = sts.chi2(2)
sample = chi2_rv.rvs(1000)
```

Построем гистограмму выборки и нарисуем поверх неё теоретическую плотность распределения нашей случайной величины:

```python
plt.hist(sample, normed=True)
x = np.linspace(-2,15,1000)
pdf = chi2_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf')
plt.legend()
plt.ylabel('number of samples')
plt.xlabel('$x$')
plt.show()
```

Генерируем 1000 выборок объёма 5.

```python
n = 5
sample = []
for i in range(1000):
    sample.append(chi2_rv.rvs(n))
```

Находим выборочное среднее каждой выборки.

```python
X = []
for i in sample:
    number = 0
    for j in i:
        number += j
    number /= float(n)
    X.append(number)
```

Мы используем распределение хи-квадрат $χ_k^2$, где в нашем случае k = 2. Из Википедии матожидание равно k = 2, дисперсия равна 2k = 4.

Посчитаем значения параметров нормальных распределений, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних, и зададим нормально распределенную случайную величину с этими параметрами.

```python
from math import sqrt
mu = 2
sigma = sqrt(4 / float(n))
norm_rv = sts.norm(loc=mu, scale=sigma)
```

Построим гистограмму распределений выборочных средних выборок и нарисуем поверх неё плотность соответствующего нормального распределения.

```python
plt.hist(X, normed=True)
x = np.linspace(-2,5,1000)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='pdf')
plt.legend()
plt.ylabel('number of samples')
plt.xlabel('$x$')
plt.show()
```
